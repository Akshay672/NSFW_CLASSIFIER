```python
#The NSFW Classifier is used to Classify Not safe for work images vs Safe images. 
#NSFW images includes Porn and Sexy images. Rest are classified as Safe Images
```

```python
from keras.models import load_model
model = load_model('Final_weights.h5')
```

```python
# https://stackoverflow.com/a/53403805/7437264
from PIL import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from skimage import transform
def load(filename):
    np_image = Image.open(filename)
    np_image = np.array(np_image).astype('float32')/255
    np_image = transform.resize(np_image, (224, 224, 3))
    np_image = np.expand_dims(np_image, axis=0)
    img=mpimg.imread(filename)
    plt.imshow(img)
    return np_image
```

```python
image = load("D://Gloify//21-01-2022//NSFW-Detection-DL-master//1.jpg")
ans = model.predict(image)
maping = {0 : "Neutral", 1 : "Porn", 2 : "Sexy"}
new_ans = np.argmax(ans[0])

print(maping[new_ans], np.round(ans,2))
print("With {} probability".format(ans[0][new_ans]))
```

```python
print("Overall the pic is identified as : ", (maping[new_ans]))
```
